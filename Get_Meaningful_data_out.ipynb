```python
import json
import pandas as pd
from pprint import pprint

symbols = pd.read_csv('companylist.csv',usecols=['Symbol'])
symbols = symbols['Symbol'].tolist()
#print(symbols)
#print("\n\n")

all_tweets = dict()
for i in range(0, len(symbols)):
    all_tweets[symbols[i]] = dict()
    all_tweets[symbols[i]]['tweets'] = list()
    all_tweets[symbols[i]]['created_at'] = list()
#pprint(all_tweets)
#print ('\n')


count = 0
with open('tweet_sample.json') as f:
    for line in f:
        data = json.loads(line)
        for i in range(0, len(data)):
            for key in data[i].keys():
                if key == "symbols":
#                     print(data[i]['symbols'])
                    for j in range(0, len(data[i]['symbols'])):
                        for kkey in data[i]['symbols'][j].keys():
                            if kkey == "symbol":
                                if data[i]['symbols'][j][kkey] in symbols:
                                    all_tweets[data[i]['symbols'][j][kkey]]['tweets'].append(data[i]['body'])
                                    all_tweets[data[i]['symbols'][j][kkey]]['created_at'].append(data[i]['created_at'])
#                                 print(data[i]['symbols'][j][kkey])
#                     print(data[i]['body'])
#                     print(data[i]['created_at'])
#                     print('\n\n')
                    
                    #pprint(data[i])
                    count +=1

pprint(all_tweets)
print(count)
```
