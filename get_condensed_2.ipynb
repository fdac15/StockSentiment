```python
import json, csv
from pprint import pprint
import pandas as pd


symbols = pd.read_csv('companylist.csv',usecols=['Symbol'])
symbols = symbols['Symbol'].tolist()

all_dates = dict()

all_tweets = dict()
for i in range(0, len(symbols)):
    all_tweets[symbols[i]] = dict()

with open('all_tweets_out_condensed.json') as data_file:        
    data = json.load(data_file)

#pprint(data)

for key in data.keys():
    for ikey in data[key].keys():
        if ikey not in all_dates:
            all_dates[ikey] = 0

        date = ikey
        total_score = (float(data[key][ikey]["positive"]) * 1) + (int(data[key][ikey]["negative"]) * -1) 
        sentiment_for_day = int(total_score)
        
        all_tweets[key][date] = sentiment_for_day          

                
for key in data.keys():
    for all_dates_key in all_dates.keys():
        if all_dates_key not in all_tweets[key]:
            all_tweets[key][all_dates_key] = all_dates[all_dates_key]                

pprint (all_tweets)


datas = pd.DataFrame.from_dict(all_tweets)

pd.DataFrame.to_csv(datas, "all_tweet.csv")
```
